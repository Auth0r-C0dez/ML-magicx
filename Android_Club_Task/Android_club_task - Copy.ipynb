Filling the income column with the mean values of the column

```python
from sklearn.impute import SimpleImputer

# Assuming you have a Pandas DataFrame `df` with a column `col_name` containing missing values
imputer = SimpleImputer(strategy='mean')  # Set the imputation strategy to mean

# Fit the imputer to the column
imputer.fit(df[["Income"]])

# Transform the column to fill missing values with the mean
filled_col = imputer.transform(df[["Income"]])

# Replace the original column with the filled values
df["Income"] = filled_col.ravel()
```

```python
import os
os.environ['MKL_THREADING_LAYER'] = 'GNU'
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Elbow Method to determine K
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show();
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    silhouette_scores.append(silhouette_score(df, kmeans.labels_))

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show();
```

```python
import warnings
from sklearn.cluster import KMeans

# Suppress specific warning related to MKL
warnings.filterwarnings("ignore", message=".*memory leak.*")
# Apply K-Means with the chosen number of clusters, e.g., K=3
kmeans = KMeans(n_clusters=4, random_state=42,n_init=10)
kmeans.fit(df)

# Get the labels (cluster assignments) and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Apply PCA for 2D visualization
pca = PCA(n_components=2)
features_pca = pca.fit_transform(df)

# Plot the clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features_pca[:, 0], y=features_pca[:, 1], hue=labels, palette='Set1')
plt.title('Customer Segments - K-Means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
# Silhouette Score
score = silhouette_score(df, labels)
print(f'Silhouette Score: {score}')
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python


bic_scores = []
aic_scores = []
n_components_range = range(1, 11)  # Try 1 to 10 components

for n in n_components_range:
    gmm = GaussianMixture(n_components=n, random_state=42)
    gmm.fit(df)
    bic_scores.append(gmm.bic(df))
    aic_scores.append(gmm.aic(df))

# Plot BIC and AIC scores
plt.plot(n_components_range, bic_scores, label='BIC')
plt.plot(n_components_range, aic_scores, label='AIC')
plt.xlabel('Number of Components')
plt.ylabel('Score')
plt.title('BIC and AIC Scores for GMM')
plt.legend()
plt.show()
```

```python
gmm = GaussianMixture(n_components=4, random_state=42)  # Replace 3 with your chosen number of clusters
gmm.fit(df)
labels = gmm.predict(df)
```

```python
silhouette_avg = silhouette_score(df, labels)
print(f'Silhouette Score for GMM: {silhouette_avg}')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Reduce dimensions for visualization
pca = PCA(n_components=2)
reduced_features = pca.fit_transform(df)  # Shape: (2240, 2)

# Ensure labels are 1D and align with reduced_features
if len(labels.shape) > 1:
    labels = labels.flatten()  # Convert to shape (2240,)

# Verify shapes match
assert reduced_features.shape[0] == labels.shape[0], "Mismatch between rows in features and labels."

# Create a DataFrame for visualization
df_viz = pd.DataFrame(reduced_features, columns=['PC1', 'PC2'])  # Shape: (2240, 2)
df_viz['Cluster'] = labels  # Add cluster labels as a column (2240,)

# Plot clusters
sns.scatterplot(data=df_viz, x='PC1', y='PC2', hue='Cluster', palette='viridis')
plt.title('GMM Clustering Results')
plt.show()
```
