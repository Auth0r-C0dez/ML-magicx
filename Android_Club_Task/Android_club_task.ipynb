Filling the income column with the mean values of the column

```python
from sklearn.impute import SimpleImputer

# Assuming you have a Pandas DataFrame `df` with a column `col_name` containing missing values
imputer = SimpleImputer(strategy='mean')  # Set the imputation strategy to mean

# Fit the imputer to the column
imputer.fit(df[["Income"]])

# Transform the column to fill missing values with the mean
filled_col = imputer.transform(df[["Income"]])

# Replace the original column with the filled values
df["Income"] = filled_col.ravel()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Initialize the MinMaxScaler
scaler_features = MinMaxScaler()
scaler_price = MinMaxScaler()

# Normalize the features
features_to_normalize = ['Year_Birth' , 'Income' ,'MntWines' , 'MntMeatProducts' , 'MntSweetProducts' ,'MntGoldProds' , 'MntFruits' , 'MntFishProducts' , 'MntGoldProds']
df[features_to_normalize] = scaler_features.fit_transform(df[features_to_normalize])
```

```python
import os
os.environ['OMP_NUM_THREADS'] = '9'
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
# Elbow Method to determine K
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.show();
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df)
    silhouette_scores.append(silhouette_score(df, kmeans.labels_))

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.show();
```

```python
import warnings
from sklearn.cluster import KMeans

# Suppress specific warning related to MKL
warnings.filterwarnings("ignore", message=".*memory leak.*")
# Apply K-Means with the chosen number of clusters, e.g., K=3
kmeans = KMeans(n_clusters=4, random_state=42,n_init=10)
kmeans.fit(df)

# Get the labels (cluster assignments) and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
# Apply PCA for 2D visualization
pca = PCA(n_components=2)
features_pca = pca.fit_transform(df)

# Plot the clusters 
plt.figure(figsize=(8, 6))
sns.scatterplot(x=features_pca[:, 0], y=features_pca[:, 1], hue=labels, palette='Set1')
plt.title('Customer Segments - K-Means Clustering')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
# Silhouette Score
score = silhouette_score(df, labels)
print(f'Silhouette Score: {score}')
```
