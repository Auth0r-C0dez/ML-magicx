```python
#Spliting x and y
x = df.drop("Country of Origin for Sample",axis =1)
y=df["Country of Origin for Sample"]

np.random.seed(42)
x_train,x_test,y_train,y_test = train_test_split(x,
                                                 y,
                                                 test_size = 0.2)
```

```python
x_train['Food Description'] = x_train['Food Description'].replace('-','',inplace=True)

```

```python
x_test['Food Description'] = x_test['Food Description'].replace('-','',inplace=True)
```

```python
x_train.info()
```

```python
#this cell was run after completion of the project
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
x_train["Category"] = x_train["Category"].astype(str).fillna("missing")
x_test["Category"] = x_test["Category"].astype(str).fillna("missing")

# Define categorical features and transformer
categorical_features = ["Category"]
one_hot = OneHotEncoder(handle_unknown='ignore')

transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)],
                                remainder="passthrough")

# Transform train and test data
transformed_X_train = transformer.fit_transform(x_train)
transformed_X_test = transformer.transform(x_test)

print("Transformed X_train:")
print(transformed_X_train)

print("\nTransformed X_test:")
print(transformed_X_test)
```

```python
#\ this cell was run after completion of the project
categorical_features=["Food Description"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",
                                  one_hot,
                                  categorical_features)],
                                remainder="passthrough")

#fill train and test values separately
x_train= transformer.fit_transform(x_train)
X_test = transformer.transform(x_test)
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(x_train, y_train);
```

```python
# Put models in a dictionary
models = {"KNN": KNeighborsClassifier(),
          "Logistic Regression": LogisticRegression(max_iter=100), # Note: if you see a warning about "convergence not reached", you can increase `max_iter` until convergence is reached
          }

# Create function to fit and score models
def fit_and_score(models, x_train, x_test, y_train, y_test):
    """
    Fits and evaluates given machine learning models.
    models : a dict of different Scikit-Learn machine learning models
    X_train : training data
    X_test : testing data
    y_train : labels assosciated with training data
    y_test : labels assosciated with test data
    """
    # Random seed for reproducible results
    np.random.seed(42)
    # Make a list to keep model scores
    model_scores = {}
    # Loop through models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(x_train, y_train)
        # Evaluate the model and append its score to model_scores
        model_scores[name] = model.score(x_test, y_test)
    return model_scores
```

```python
model_scores = fit_and_score(models=models,
                             x_train=x_train,
                             x_test=x_test,
                             y_train=y_train,
                             y_test=y_test)
model_scores
```
